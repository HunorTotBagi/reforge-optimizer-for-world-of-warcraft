```python
import itertools

class Item:
    stats = ['hit', 'expertise', 'crit', 'haste', 'mastery']

    def __init__(self, name, stats):
        self.name = name
        self.stats = stats

    def reforge(self, source_stat, target_stat):
        amount_to_reforge = int(self.stats[source_stat] * 0.4)
        self.stats[source_stat] -= amount_to_reforge
        self.stats[target_stat] += amount_to_reforge

    def get_combinations(self):
        combinations = []
        for r in range(1, len(Item.stats) + 1):
            for combination in itertools.combinations(Item.stats, r):
                combinations.append(combination)
        return combinations

    def print_combinations(self):
        combinations = self.get_combinations()
        for combination in combinations:
            print(f"Item: {self.name}")
            print(f"Combination: {combination}")
            for source_stat in combination:
                remaining_stats = set(Item.stats) - set(combination)
                for target_stat in remaining_stats:
                    item_clone = Item(self.name, self.stats.copy())
                    item_clone.reforge(source_stat, target_stat)
                    print(f"  Reforging {source_stat} into {target_stat}: {item_clone.stats}")
            print()

# Example usage
head_piece = Item("Head", {'hit': 100, 'expertise': 200, 'crit': 300, 'haste': 400, 'mastery': 500})
necklace = Item("Neck", {'hit': 50, 'expertise': 150, 'crit': 250, 'haste': 350, 'mastery': 450})
# ... Add more items here

head_piece.print_combinations()
necklace.print_combinations()
# ... Print combinations for other items
```
